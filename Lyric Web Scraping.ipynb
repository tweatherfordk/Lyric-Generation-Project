# Corpus Creation

```python
def LyricDataframe(artist):
   
    #initialize empty attribute lists
    artist_name = str(artist.name)
    artist_name_list = []
    title = []
    lyrics = []
    
    #append song attributes to lits
    for song in artist.songs:
        artist_name_list.append(artist_name)
        title.append(song.title)
        lyrics.append(song.lyrics)
    
    #format lists to feed into dataframe
    key_list = ['Artist', 'Song Title', 'Lyrics']
    value_list = [artist_name_list, title, lyrics]
    
    #create dataframe for artist
    df = pd.DataFrame(value_list).T.set_axis(key_list, axis=1, inplace=False)
    return df
```

```python
#apply function to each band
lifetime_df = LyricDataframe(lifetime)
cure_df = LyricDataframe(cure)
jawbreaker_df = LyricDataframe(jawbreaker)
mcr_df = LyricDataframe(mcr)
blink_df = LyricDataframe(blink)
sunnyday_df = LyricDataframe(sunnyday)
gerard_df = LyricDataframe(gerard)
```

```python
#create corpus
corpus_df = pd.concat([lifetime_df, cure_df, jawbreaker_df, mcr_df, blink_df, sunnyday_df, gerard_df]).reset_index().drop(['index'], axis=1)
corpus_df
```
